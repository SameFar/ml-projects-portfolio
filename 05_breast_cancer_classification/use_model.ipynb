```python
import pandas as pd

# Im just using the test/training data
df = pd.read_csv("data.csv")
df = df.drop(
    columns=[
        "id",
        "texture_se",
        "smoothness_se",
        "symmetry_se",
        "fractal_dimension_se",
        "fractal_dimension_mean",
    ]
)
df = df.sample(1)
X = df.drop(columns="diagnosis")
```

### Load all models

```python
import pickle

models_dict = {
    "SVC": None,
    "KNN": None,
    "Logistic Regression": None,
    "Naive Bayes": None,
    "Random Forest": None,
    "Decision Tree": None,
    "XGBoost": None,
    "SGD": None,
}

for name in models_dict.keys():
    filename = f"{name}Model.pkl"
    with open(filename, "rb") as f:
        models_dict[name] = pickle.load(f)
```

### Predictions

```python
predictions = {
    "SVC": None,
    "KNN": None,
    "Logistic Regression": None,
    "Naive Bayes": None,
    "Random Forest": None,
    "Decision Tree": None,
    "XGBoost": None,
    "SGD": None,
}
for name, model in models_dict.items():
    predictions[name] = model.predict(X)[0]
```

### Display results

```python
vote = sum(1 for value in predictions.values() if value == 1)

print("Decision:", "Malignant" if vote > 3 else "Benige")
print(f"\nPrediction: {vote}/8 Malignant\n")

for model, pred in predictions.items():
    print(f"{model}: {pred}")
```
