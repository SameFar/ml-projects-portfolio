```python
import yfinance as yf
import numpy as np
import pandas as pd
```

```python
data = yf.download("NVDA", start="2026-01-01", end="2026-05-01")
```

```python
data.isna().sum()
```

```python
df = pd.DataFrame()
df['Volitility'] = np.log(data['High']/data['Low'])
df['Log_Returns'] = np.log(data['Close'] / data['Close'].shift(1))
df['Volume'] = (data['Volume']-data['Volume'].min())/(data['Volume'].max()-data['Volume'].min()) # Min-Max Scale
df['Next_Open'] = (data['Open'].shift(-1) - data['Open'].min()) / (data['Open'].max() - data['Open'].min())
```

```python
df.dropna(inplace=True)
```

```python
df.to_csv("nvidea_stocks2.csv")
```
