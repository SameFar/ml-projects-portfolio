```python
import yfinance as yf
import numpy as np
import pandas as pd
```

```python
data = yf.download("NVDA", start="2026-01-01", end="2026-06-01")
```

```python
df = pd.DataFrame()

df['Volatility'] = np.log(data['High'] / data['Low'])
df['Log_Returns'] = np.log(data['Close'] / data['Close'].shift(1))
df['Volume_Scaled'] = data['Volume'] / data['Volume'].rolling(window=20).mean() - 1

# Target
df['Close'] = data['Close']
```

```python
df.dropna(inplace=True)
```

```python
df.to_csv("nvidea_stocks_pred.csv")
```
